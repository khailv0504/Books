```python
import torch
import torch.nn as nn
import kagglehub
from ptflops import get_model_complexity_info
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split
from contextlib import contextmanager
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import time
import numpy as np
from torchvision.models import googlenet, GoogLeNet_Weights

class EMA:
    def __init__(self, model, decay=0.999):
        self.shadow = {n: p.data.clone() for n, p in model.named_parameters() if p.requires_grad}
        self.backup = {}
        self.decay = decay

    def update(self, model):
        with torch.no_grad():
            for name, param in model.named_parameters():
                if param.requires_grad:
                    # Equation: shadow = decay * shadow + (1 - decay) * current
                    self.shadow[name].copy_(self.decay * self.shadow[name] + (1.0 - self.decay) * param.data)

    @contextmanager
    def average_parameters(self, model):
        """Context manager temporalily apply to weights EMA for validation"""
        # Lưu lại trọng số đang training
        for name, param in model.named_parameters():
            if param.requires_grad:
                self.backup[name] = param.data.clone()
                param.data.copy_(self.shadow[name])
        try:
            yield
        finally:
            # Restore original weights in order to training
            for name, param in model.named_parameters():
                if param.requires_grad:
                    param.data.copy_(self.backup[name])
            self.backup.clear()


def evaluate(model, val_loader, device):

    model.eval()
    criterion = torch.nn.CrossEntropyLoss().to(device)
    correct, total, loss_sum = 0, 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            losses = criterion(outputs, labels)
            loss_sum += losses.item() * inputs.shape[0]
            correct += (outputs.argmax(dim=1) == labels).sum().item()
            total += inputs.shape[0]
    return loss_sum / total, correct / total


def train_model(
        model,
        train_loader,
        val_loader,
        device,
        num_epochs=25
):

    train_loss_history = []
    val_loss_history = []
    train_acc_history = []
    val_acc_history = []

    model.to(device)

    # Define optimizer, scheduler, losses function, scaler, ema
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=0.001,
        steps_per_epoch=len(train_loader),
        epochs=num_epochs,
        pct_start=0.2,
        anneal_strategy="cos"
    )
    criterion = torch.nn.CrossEntropyLoss(label_smoothing=0.1).to(device)
    scaler = torch.amp.GradScaler("cuda")
    ema = EMA(model)

    # Best accuracy on validation set
    best_acc = 0.0
    best_ema_weights = None

    start_time = time.time()
    for epoch in range(1, num_epochs + 1):
        model.train()
        total_loss, correct, total = 0, 0, 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}")
        for inputs, labels in pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            # Using amp in order to optimize VRAM GPU and speed training
            with torch.amp.autocast("cuda"):
                outputs = model(inputs)
                losses = criterion(outputs, labels)
            # Backward
            scaler.scale(losses).backward()
            scaler.step(optimizer)
            scaler.update()
            scheduler.step()
            ema.update(model)

            total_loss += losses.item() * inputs.shape[0]
            correct += (outputs.argmax(dim=1) == labels).sum().item()
            total += inputs.shape[0]
            pbar.set_postfix(loss=f"{total_loss / total:.4f}", acc=f"{(correct / total) * 100:.2f}%")

        # Evaluate on validation set
        with ema.average_parameters(model):
            # Lúc này model đang dùng trọng số EMA
            val_loss, val_acc = evaluate(model, val_loader, device)
        # Sau block này, model tự động quay về trọng số gốc
        train_acc = correct / total
        train_loss = total_loss / total

        train_loss_history.append(train_loss)
        val_loss_history.append(val_loss)
        train_acc_history.append(train_acc)
        val_acc_history.append(val_acc)

        # Print currently additional learning rate to track
        current_lr = optimizer.param_groups[0]['lr']
        print(f"Epoch {epoch}: TrainAcc={train_acc * 100:.2f}%, ValAcc={val_acc * 100:.2f}% | LR={current_lr:.6f}")

        if val_acc > best_acc:
            best_acc = val_acc
            best_ema_weights = {k: v.clone() for k, v in ema.shadow.items()}
            print(f"🌟 Best model saved (ValAcc={best_acc * 100:.2f}%)")
    end_time = time.time()

    # Reload best EMA weights
    for name, param in model.named_parameters():
        if param.requires_grad:
            param.data.copy_(best_ema_weights[name])

    print(f"\nTraining done. Best ValAcc={best_acc * 100:.2f}%")

    # Compute training time
    elapsed = end_time - start_time
    hours = int(elapsed // 3600)
    minutes = int((elapsed % 3600) // 60)
    seconds = int(elapsed % 60)
    print(f"Training time: {hours}h {minutes}m {seconds}s")

    return train_loss_history, val_loss_history, train_acc_history, val_acc_history

def draft_curve(model,
                val_loader,
                train_loss_history,
                val_loss_history,
                train_acc_history,
                val_acc_history,
                class_names):
    epochs = range(1, len(train_loss_history) + 1)

    # ===== LOSS =====
    plt.figure()
    plt.plot(epochs, train_loss_history, label="Train Loss")
    plt.plot(epochs, val_loss_history, label="Validation Loss")

    # ===== FIND MIN =====
    min_train_loss = min(train_loss_history)
    min_train_epoch = np.argmin(train_loss_history) + 1

    min_val_loss = min(val_loss_history)
    min_val_epoch = np.argmin(val_loss_history) + 1

    # ===== ANNOTATE =====
    plt.scatter(min_train_epoch, min_train_loss, color='blue')
    plt.scatter(min_val_epoch, min_val_loss, color='orange')

    plt.annotate(f"{min_train_loss:.4f}",
                 (min_train_epoch, min_train_loss),
                 xytext=(0, 10),
                 textcoords="offset points",
                 ha='center')

    plt.annotate(f"{min_val_loss:.4f}",
                 (min_val_epoch, min_val_loss),
                 xytext=(0, 10),
                 textcoords="offset points",
                 ha='center')

    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Biểu đồ training loss")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ===== ACCURACY =====
    plt.figure()
    plt.plot(epochs, train_acc_history, label="Train Accuracy")
    plt.plot(epochs, val_acc_history, label="Validation Accuracy")

    # ===== FIND MAX =====
    max_train_acc = max(train_acc_history)
    max_train_epoch = np.argmax(train_acc_history) + 1

    max_val_acc = max(val_acc_history)
    max_val_epoch = np.argmax(val_acc_history) + 1

    # ===== PLOT POINT =====
    plt.scatter(max_train_epoch, max_train_acc, color='blue')
    plt.scatter(max_val_epoch, max_val_acc, color='orange')

    # ===== ANNOTATE =====
    plt.annotate(f"{max_train_acc:.4f}",
                 (max_train_epoch, max_train_acc),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')

    plt.annotate(f"{max_val_acc:.4f}",
                 (max_val_epoch, max_val_acc),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha='center')

    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Biểu đồ accuracy")
    plt.legend()
    plt.grid(True)
    plt.show()

    # ===== EVALUATION =====
    all_preds = []
    all_labels = []

    # Uses global val_loader and device as noted in the docstring.
    model.eval()
    model.to(device)

    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())


    # ===== CONFUSION MATRIX =====
    cm = confusion_matrix(all_labels, all_preds)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d",
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    # ===== NORMALIZED CM =====
    cm_norm = cm.astype("float") / cm.sum(axis=1)[:, None]

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_norm, annot=True, fmt=".2f",
                xticklabels=class_names,
                yticklabels=class_names)
    plt.title("Normalized Confusion Matrix")
    plt.show()

    # ===== REPORT =====
    print(classification_report(
        all_labels,
        all_preds,
        target_names=class_names
    ))

def cnt_time_inference(model, device, val_loader, batch_size=32):
    model.eval()
    device = torch.device(device)

    print("Running warm-up...")
    with torch.no_grad():
        inputs, _ = next(iter(val_loader))
        inputs = inputs.to(device)
        _ = model(inputs)
        if torch.cuda.is_available():
            torch.cuda.synchronize()

    print("Measuring inference speed...")
    total_time = 0.0
    num_batches = len(val_loader)

    with torch.no_grad():
        for inputs, _ in tqdm(val_loader, desc="Predicting"):
            inputs = inputs.to(device)

            start = time.time()
            _ = model(inputs)

            if torch.cuda.is_available():
                torch.cuda.synchronize()

            end = time.time()
            total_time += (end - start)

    avg_batch_time_ms = (total_time / num_batches) * 1000
    avg_img_time_ms = avg_batch_time_ms / batch_size
    fps = 1000 / avg_img_time_ms

    print("\n=======================")
    print(f"Total inference time       : {total_time*1000:.2f} ms")
    print(f"Average time per batch     : {avg_batch_time_ms:.3f} ms")
    print(f"Average time per image     : {avg_img_time_ms:.3f} ms")
    print(f"FPS (frames per second)    : {fps:.2f}")
    print("=======================\n")


if __name__ == "__main__":
    # Download laval version
    path = kagglehub.dataset_download("huynhthethien/radarcommunsignaldata2026train")
    print("Path to dataset files:", path)
    data_root = path

    # Thêm ColorJitter và RandomRotation được thầy cho phép
    train_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ColorJitter(brightness=0.2, contrast=0.2),  # [Cho phép: mục 15.1.4]
        transforms.ToTensor(),
        transforms.RandomErasing(p=0.5, scale=(0.02, 0.1), ratio=(0.1, 3.3), value=0)
    ])

    val_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])

    # Gán biến đổi tách biệt cho train và val
    full_dataset_train = datasets.ImageFolder(data_root, transform=train_transform)
    full_dataset_val = datasets.ImageFolder(data_root, transform=val_transform)

    train_size = int(0.8 * len(full_dataset_train))
    val_size = len(full_dataset_train) - train_size

    generator = torch.Generator().manual_seed(42)
    train_ds, _ = random_split(full_dataset_train, [train_size, val_size], generator=generator)
    _, val_ds = random_split(full_dataset_val, [train_size, val_size], generator=generator)

    train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_ds, batch_size=32, shuffle=False, num_workers=2)

    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Apply for googLeNet
    model = googlenet(weights=GoogLeNet_Weights.DEFAULT)
    in_features = model.fc.in_features
    model.fc = nn.Linear(in_features, 12)
    model = model.to(device)

    # ===== FLOPs / MACs =====
    with torch.cuda.device(0):
        macs, params = get_model_complexity_info(
            model,
            (3, 224, 224),
            as_strings=False,
            print_per_layer_stat=False,
            verbose=False
        )

    print(f"FLOPs : {macs*2 / 1e6:.2f} M")
    print(f"MACs  : {macs / 1e6:.2f} M")
    print(f"Params: {params:,}  ({params / 1e3:.3f} K, {params / 1e6:.4f} M)")
    cnt_time_inference(model, device, val_loader, batch_size=32)

    train_loss_history, val_loss_history, train_acc_history, val_acc_history = train_model(model, train_loader, val_loader, device)
    draft_curve(model, val_loader, train_loss_history, val_loss_history, train_acc_history, val_acc_history, full_dataset_train.classes)
    # export torchscript model
    model.eval()
    example_input = torch.randn(1, 3, 224, 224).to(device)
    # convert PyTorch model into TorchScript format
    traced_model = torch.jit.trace(model, example_input)

    # save model
    GroupID = "15"
    model_name = f"{GroupID}_DeepLearningProject_TrainedModel.pt"
    traced_model.save(model_name)
    print("Model saved:", model_name)


```
